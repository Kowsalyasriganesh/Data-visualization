```python
x = df.drop('Churn',axis = 1)
y = df['Churn']
plt.figure(figsize = (20,15))
plotnumber = 1
for col in x.columns:
  if plotnumber <= 20:
    ax = plt.subplot(5, 5, plotnumber)
    plt.plot(x[col])
    plt.xlabel(col, fontsize = 20)
  plotnumber += 1
plt.tight_layout()
plt.show()
```
